```python
road_crash_data_2020 = road_crash_data[['Month', '2020']]
```

```python
x = road_crash_data_2020['Month']
y = road_crash_data_2020['2020']
```

```python
type(x)
```

```python
x = x.tolist()
```

```python
type(x)
```

```python
x
```

```python
y = list(map(int, y))
```

```python
y
```

```python
plt.plot(x,y,marker='o')
plt.title("Road Crashes in Uganda by Month (2020)")
plt.xlabel("Months")
plt.ylabel("Number of Road Crashes")
#joins the x and y values
for x,y in zip(x,y):

    label = "{:,}".format(y)

    plt.annotate(label, # this is the value which we want to label (text)
                 (x,y), # x and y is the points location where we have to label
                 textcoords="offset points",
                 xytext=(0,10), # this for the distance between the points
                 ha='center')
plt.show()
```
